# **Matplotlib**
Matplotlib is a Python 2D plotting library which produces publication-quality figures in a variety of hardcopy formats and interactive environments across platforms.

The easiest way to use Matplotlib to follow the following steps systematically

---
## **Prepare The Data**
### **1 D Data**

```python
import numpy as np
x = np.linspace(0, 10, 100)
y = np.cos(x)
z = np.sin(x)
```

### **2 D Data**

```python
data = 2 * np.random.random((10, 10))
data = 2 * np.random.random((10, 10))
data2 = 3 * np.random.random((10, 10))
Y, X = np.mgrid[-3:3:100j, -3:3:100j]
U = -1 - X**2 + Y
V = 1 + X - Y**2
img = np.load(np.DataSource().open('https://github.com/matplotlib/sample_data/blob/master/axes_grid/bivariate_normal.npy?raw=true').name)
```

---
## **Create Plot**

```python
import matplotlib.pyplot as plt
```

### **Figure**

```python
fig = plt.figure()
```

```python
fig2 = plt.figure(figsize=plt.figaspect(2.0))
```

### **Axes**
All plotting is done with respect to an Axes. In most cases, a subplot will fit your needs. A subplot is an axes on a grid system.

```python
fig.add_axes()
```

```python
ax1 = fig.add_subplot(221) # row-col-num
```

```python
ax3 = fig.add_subplot(212)
```

```python
fig3, axes = plt.subplots(nrows=2,ncols=2)
```

```python
fig4, axes2 = plt.subplots(ncols=3)
```

---
## **Plotting Routines**
### **1 D Data**

```python
fig, ax = plt.subplots()
lines = ax.plot(x,y)
ax.scatter(x,y)
axes[0,0].bar([1,2,3],[3,4,5])
axes[1,0].barh([0.5,1,2.5],[0,1,2])
axes[1,1].axhline(0.45)
axes[0,1].axvline(0.65)
ax.fill(x,y,color='blue')
ax.fill_between(x,y,color='yellow')
```

### **Vector Fields**

```python
axes[0,1].arrow(0,0,0.5,0.5)
axes[1,1].quiver(y,z)
axes[0,1].streamplot(X,Y,U,V)
```

### **Data Distributions**

```python
ax1.hist(y)
ax3.boxplot(y)
ax3.violinplot(z)
```

### **2 D Data**

```python
fig, ax = plt.subplots()
im = ax.imshow(img, cmap='gist_earth', interpolation='nearest', vmin=-2, vmax=2)
axes2[0].pcolor(data2)
axes2[0].pcolormesh(data)
CS = plt.contour(Y,X,U)

#axes2[2].contourf(data1)

axes2[2]= ax.clabel(CS)
```

## **Customize Plot**
### **Colors, Color Bars & Color Maps**

```python
plt.plot(x, x, x, x**2, x, x**3)
ax.plot(x, y, alpha = 0.4)
ax.plot(x, y, c='k')
fig.colorbar(im, orientation='horizontal')
im = ax.imshow(img, cmap='seismic')
```

### **Markers**

```python
fig, ax = plt.subplots()
ax.scatter(x,y,marker=".")
ax.plot(x,y,marker="o")
```

### **Linestyles**

```python
plt.plot(x,y,linewidth=4.0)
plt.plot(x,y,ls='solid')
plt.plot(x,y,ls='--')
plt.plot(x,y,'--',x**2,y**2,'-.')
plt.setp(lines,color='r',linewidth=4.0)
```

### **Text & Annotations**

```python
ax.text(1, -2.1, 'Example Graph', style='italic')
ax.annotate("Sine",xy=(8, 0),xycoords='data',xytext=(10.5, 0),textcoords='data',arrowprops=dict(arrowstyle="->",connectionstyle="arc3"),)
```

### **Mathtext**

```python
plt.title(r'$sigma_i=15$', fontsize=20)
```

### **Limits, Legends & Layouts**
#### **Limits & Autoscaling**

```python
ax.margins(x=0.0,y=0.1)
ax.axis('equal')
ax.set(xlim=[0,10.5],ylim=[-1.5,1.5])
ax.set_xlim(0,10.5)
```

#### **Legends**

```python
ax.set(title='An Example Axes', ylabel='Y-Axis', xlabel='X-Axis')
ax.legend(loc='best')
```

#### **Ticks**

```python
ax.xaxis.set(ticks=range(1,5), ticklabels=[3,100,-12,"foo"])
ax.tick_params(axis='y', direction='inout', length=10)
```

#### **Subplot Spacing**

```python
fig3.subplots_adjust(wspace=0.5,hspace=0.3, left=0.125, right=0.9, top=0.9, bottom=0.1)
fig.tight_layout()
```

#### **Axis Spines**

```python
ax1.spines['top'].set_visible(False)
ax1.spines['bottom'].set_position(('outward',10))
```

# **Tiny Example**

```python
import matplotlib.pyplot as plt
# Prepare data
x = [1,2,3,4]
y = [10,20,25,30]
# Create plot
fig = plt.figure()
# Plot
ax = fig.add_subplot(111)
# Customize plot
ax.plot(x, y, color='lightblue', linewidth=3)
ax.scatter([2,4,6], [5,15,25], color='darkgreen', marker='^')
ax.set_xlim(1, 6.5)
# Save plot
plt.savefig('foo.png')
# Show plot
plt.show()
```
